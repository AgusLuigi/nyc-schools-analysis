```python
import os
import shutil

# ==========================================
# ⚙️ USER CONFIGURATION
# ==========================================
GITHUB_USERNAME = "AgusLuigi"
REPO_NAME = "nyc-schools-analysis"
SOURCE_DIR_NAME = "daily_tasks"  # Name of the source directory
MAX_SEARCH_DEPTH = 3             # How many levels '..' to look up?

# ==========================================
# 📂 PROJECT STRUCTURE & CONTENT
# ==========================================
project_structure = {
    "incident_analysis": {
        "readme": "# 🛡️ Incident Analysis\n\n## Task\nAnalysis of safety incidents in NYC schools (Former Day 1).\n\n## Key Findings\n- Identification of high-incident boroughs.\n- Correlation between school size and safety reports.",
        "search_pattern": "day_1"
    },
    "school_directory_exploration": {
        "readme": "# 🏫 School Directory Exploration\n\n## Task\nExploration of the NYC School Directory dataset (Former Day 2).\n\n## Insights\n- Mapping school distributions across districts.\n- Categorization of school types and specialized programs.",
        "search_pattern": "day_2"
    },
    "database_queries": {
        "readme": "# 🔍 Database Queries\n\n## Task\nExecuting complex SQL queries via Python to extract specific metrics (Former Day 3).\n\n## Outcomes\n- Joined incident data with school directories.\n- Aggregated performance metrics by borough.",
        "search_pattern": "day_3"
    },
    "database_population": {
        "readme": "# 🗄️ Database Population\n\n## Task\nData cleaning and automated upload to a SQL database (Former Day 4).\n\n## Logic\n- Handled missing values and data type conversion.\n- Used SQLAlchemy for efficient batch uploads.",
        "search_pattern": "day_4"
    },
    "assets": {
        "readme": "# 🎨 Assets\n\nThis folder contains visualizations, charts, and screenshots generated during the analysis.",
        "search_pattern": "assets"
    }
}

main_readme_content = f"""# 🍎 NYC Schools Data Analysis Portfolio

## 📌 Project Overview
This repository showcases a complete end-to-end data engineering and analysis workflow using NYC Open Data. 
The project covers everything from initial data exploration to SQL database integration and visualization.

## 🛠️ Tech Stack
- **Language:** Python
- **Libraries:** Pandas, SQLAlchemy, Matplotlib, Seaborn
- **Database:** SQLite / PostgreSQL
- **Environment:** Jupyter Notebooks

---
*Created by {GITHUB_USERNAME} as part of the Data Analytics Onboarding 2026.*
"""

# ==========================================
# 🛠️ LOGIC FUNCTIONS
# ==========================================

def find_source_directory(folder_name, max_depth):
    """Recursively searches upwards (..) to find the source folder."""
    current_path = os.getcwd()
    for i in range(max_depth + 1):
        check_path = os.path.join(current_path, folder_name)
        if os.path.exists(check_path):
            return check_path
        current_path = os.path.dirname(current_path)
    return None

def create_redirect_file():
    """Creates the required redirect Markdown file for the Day 5 submission."""
    filename = f"{GITHUB_USERNAME}_project_link.md"
    content = f"""# 📎 NYC Schools Portfolio Project

🔗 **Public GitHub Repository**: [https://github.com/{GITHUB_USERNAME}/{REPO_NAME}](https://github.com/{GITHUB_USERNAME}/{REPO_NAME})

This repository contains my cleaned, documented, and organized portfolio version of the NYC Schools Data Analysis project. It includes:
- Clean data pipelines  
- SQL integration via Python  
- Visualizations and insights  
- Well-structured documentation  

Feel free to explore!
"""
    with open(filename, "w", encoding="utf-8") as f:
        f.write(content)
    print(f"✅ Redirect file created: {filename}")
    return filename

def setup_portfolio():
    # 1. Search for source directory
    source_root = find_source_directory(SOURCE_DIR_NAME, MAX_SEARCH_DEPTH)
    
    if not source_root:
        print(f"❌ ERROR: Folder '{SOURCE_DIR_NAME}' not found even after {MAX_SEARCH_DEPTH} levels of '..'!")
        return
    
    print(f"✅ Source directory found: {source_root}")

    # 2. Create folders and READMEs
    with open("README.md", "w", encoding="utf-8") as f:
        f.write(main_readme_content)
    
    for folder, content in project_structure.items():
        if not os.path.exists(folder):
            os.makedirs(folder)
        with open(os.path.join(folder, "README.md"), "w", encoding="utf-8") as f:
            f.write(content["readme"])
        print(f"📁 Folder prepared: {folder}")

    # 3. Create requirements.txt
    if not os.path.exists("requirements.txt"):
        with open("requirements.txt", "w", encoding="utf-8") as f:
            f.write("pandas\nsqlalchemy\nmatplotlib\nseaborn\nnotebook")
        print("✅ requirements.txt created.")

    # 4. Copy files
    print("\n📂 Copying files...")
    files_count = 0
    for root, dirs, files in os.walk(source_root):
        for file in files:
            # Skip hidden files and existing READMEs
            if file.startswith('.') or file.lower() == "readme.md":
                continue
            for target_folder, info in project_structure.items():
                if info["search_pattern"] in root.lower():
                    shutil.copy2(os.path.join(root, file), os.path.join(target_folder, file))
                    print(f"  📄 {file} -> {target_folder}/")
                    files_count += 1

    # 5. Create Day 5 Redirect file
    redirect_name = create_redirect_file()

    print(f"\nSuccess! {files_count} files successfully copied.")
    print(f"Next step: Copy '{redirect_name}' into 'daily_tasks/day_5/' of your onboarding repo.")

if __name__ == "__main__":
    setup_portfolio()
```
