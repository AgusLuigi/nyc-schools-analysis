```python
#🔎 Object Collums visualize
def analyze_semantic_and_row_integrity(df, max_unique=50):
    """
    Enhanced: Identifies object classes and checks for row-level integrity 
    including metadata detection and cardinality checks.
    """
    df_check = df
    potential_classes = []
    needs_manual_review = []
    
    print("SEMANTIC & ROW INTEGRITY ANALYSIS")
    
    for col in df_check.columns:
        # Preparation
        unique_vals = df_check[col].dropna().unique()
        unique_count = len(unique_vals)
        total_count = len(df_check[col])
        null_count = df_check[col].isnull().sum()
        nan_percentage = (null_count / total_count) * 100
        cardinality_ratio = (unique_count / total_count) * 100
        
        # 1. Basic Identification of Category/Object Class
        # Added: Check if it's not just a single repeated value (unique_count > 1)
        is_category = (1 < unique_count <= max_unique) or ('name' in col.lower() or 'dbn' in col.lower())
        
        if is_category:
            row_issues = []
            
            # A. Check for NaNs (Critical if > 90%)
            if null_count > 0:
                severity = "CRITICAL" if nan_percentage > 90 else "Warning"
                row_issues.append(f"{null_count} NaNs ({nan_percentage:.1f}%) [{severity}]")
            
            # B. Check for Mixed Types
            types_in_col = df_check[col].apply(type).unique()
            if len(types_in_col) > 1:
                row_issues.append(f"Mixed types: {list(types_in_col)}")
            
            # C. PLUS: Metadata Detection (URLs, Emails, Phone Numbers)
            if unique_count > 0:
                sample_str = str(unique_vals[0]).lower()
                patterns = {
                    'URL': r'http[s]?://',
                    'Email': r'[^@]+@[^@]+\.[^@]+',
                    'Phone': r'\d{3}-\d{3}-\d{4}'
                }
                for label, pattern in patterns.items():
                    if re.search(pattern, sample_str):
                        row_issues.append(f"Technical Metadata ({label}) detected")

            # D. PLUS: High Cardinality Check (Identifying hidden IDs)
            if cardinality_ratio > 95 and 'name' not in col.lower():
                row_issues.append(f"High Cardinality ({cardinality_ratio:.1f}%): Likely an ID, not a Class")

            # E. String Length Check
            if df_check[col].dtype == 'object' and unique_count <= max_unique:
                avg_len = pd.Series([len(str(x)) for x in unique_vals]).mean()
                if avg_len > 100:
                    row_issues.append("Text content too long for visualization labels")

            # Final Decision Logic
            if row_issues:
                needs_manual_review.append({
                    'column': col,
                    'unique_count': unique_count,
                    'issues': row_issues
                })
            else:
                potential_classes.append(col)

    return potential_classes, needs_manual_review

# EXECUTION
suggested_objects, review_list = analyze_semantic_and_row_integrity(df)

print("✅ SAFE OBJECT CLASSES (Clean rows & names):")
print(suggested_objects)

if review_list:
    print("⚠️ COLUMNS REQUIRING ROW-LEVEL CLEANING OR REVIEW:")
    for item in review_list:
        issues_str = " | ".join(item['issues'])
        print(f"- {item['column']} ({item['unique_count']} uniques): {issues_str}")
```

```python
# Goal: Count unique schools in Brooklyn
# Filter: borough == 'Brooklyn'
def check_total_brooklyn_schools(df):
    brooklyn_data = df[df['borough'].str.contains('Brooklyn', case=False, na=False)]
    total_unique = brooklyn_data['dbn'].nunique()
    return total_unique

total_count = check_total_brooklyn_schools(df)
print(f"Total number of unique schools in Brooklyn: {total_count}")
```

```python
# Filters for borough 'Brooklyn' and an exact start at grade 9
def cross_check_brooklyn_9(df):
    """
    Direct calculation based on the data structure:
    Filters for borough 'Brooklyn' and an exact start at grade 9.
    """
    # 1. Filter: borough must be Brooklyn
    # 2. Filter: grade_span_min must be 9 (entry starting from grade 9 only)
    result_df = df[
        (df['borough'].str.contains('Brooklyn', case=False, na=False)) & 
        (df['grade_span_min'] >= 9)
    ]
    count = len(result_df)
    school_names = result_df['school_name'].tolist()
    return count, school_names

# Execution
count, names = cross_check_brooklyn_9(df)
print(f"Number of schools: {count}")
#print(f"Identified schools: {names}")
```

```python
# # Unique count per borough
def calculate_schools_per_borough(df):
    # Unique count per borough
    stats = df.groupby('borough')['dbn'].nunique().reset_index()
    stats.columns = ['Borough', 'Number_of_Schools']
    
    # Sort for better overview
    stats = stats.sort_values(by='Number_of_Schools', ascending=False)
    
    # Validation: sum must equal 435 (according to your shape report)
    total_check = stats['Number_of_Schools'].sum()
    
    return stats, total_check

boroughs_df, total_sum = calculate_schools_per_borough(df)
print(boroughs_df.to_string(index=False))
print(f"\nTotal sum verified: {total_sum} (Expected: 435)")
```

```python
# Grouping and calculating the mean
def student_stats_per_borough(df):
    # Grouping and calculating the mean
    stats = df.groupby('borough')['total_students'].mean().round(2).reset_index()
    stats.columns = ['Borough', 'Average_Student_Count']
    
    # Sort by size
    stats = stats.sort_values(by='Average_Student_Count', ascending=False)
    
    return stats

# Execution
avg_students_df = student_stats_per_borough(df)
print("Average student count per borough:")
print(avg_students_df.to_string(index=False))
```

```python
# Statistical summary of grade_span_max by borough
def grade_summary_per_borough(df):
    # The describe() method returns count, mean, std, min, 25%, 50%, 75%, max
    summary = df.groupby('borough')['grade_span_max'].describe()
    
    # Optional: Rename columns for English output
    summary.columns = ['Count', 'Mean', 'StdDev', 'Min', '25%', 'Median', '75%', 'Max']
    
    return summary

# Execution
grade_stats = grade_summary_per_borough(df)
print("Statistical summary of grade_span_max by borough:")
print(grade_stats)
```
